# Steps in Evaluating AI Models

---

### 1️⃣ Define Success Criteria

Before evaluating a model, you must decide what “good” means.
For example:

* In classification: you might define success as “at least 90% accuracy.”
* In text generation: you might value “fluency” or “relevance.”
* In fairness or safety contexts: success means avoiding biased or harmful responses.

This step ensures your evaluation aligns with the *real goal* — not just a number but the intended outcome (usefulness, reliability, or safety).

---

### 2️⃣ Prepare a Ground Truth Dataset

This means assembling a **set of examples with correct answers** — the reference standard against which the model’s outputs will be compared.
The dataset can be created by:

* Collecting real labeled data (e.g., images with correct categories).
* Having experts or annotators label examples manually.
* Using existing benchmark datasets if they fit your task.

This is the “gold standard” — the truth that lets you tell whether your model is right or wrong.

---

### 3️⃣ Run the Model to Generate Predictions

Once you have input examples and known correct answers, run the model to produce its outputs or predictions.
For example:

* Feed test images into a classifier.
* Ask a text model to summarize an article.
* Predict prices for unseen housing data.

This gives you the model’s *version of the truth* — what it thinks the answer should be.

---

### 4️⃣ Compare Predictions with Ground Truth

Now measure how close the model’s outputs are to the ground truth.
Depending on the task, this could mean:

* Checking whether predicted labels match the true labels (accuracy).
* Measuring how close numeric predictions are (mean squared error).
* Assessing similarity between generated and reference text (BLEU, ROUGE, etc.).

This step quantifies performance — it’s how you convert “right or wrong” into metrics you can track.

---

### 5️⃣ Aggregate Results for Consistency and Reliability

A single example doesn’t tell the full story.
You combine all the individual results to see **overall performance trends**:

* Average accuracy across all samples.
* Distribution of errors (are some categories much worse?).
* Stability across different subsets or conditions.

This helps you judge not only how *good* the model is but how *consistent* and *trustworthy* it remains across scenarios.

---

### 💡 In Summary

The evaluation process is essentially a **scientific test cycle**:

1. Decide what success means.
2. Establish trusted “correct” answers.
3. Run the model on unseen data.
4. Measure how well predictions match reality.
5. Summarize and interpret results for reliability.

Together, these steps ensure your model isn’t just producing outputs — it’s producing *useful, verifiable, and repeatable* results that meet your real-world goals.


# Does evaluation happen only after model training?

No, as you will read below.  BUT, the focus of this class is on evaluating models after training!
---

### 🧩 During Training

Evaluation often begins **while the model is still being trained**, using a **validation dataset**.
At this stage, you’re not just looking for the final accuracy — you’re watching *learning progress*.
You evaluate repeatedly to check:

* Whether the model is improving over time.
* Whether it’s overfitting (memorizing training data instead of learning patterns).
* When to stop training (early stopping based on validation performance).

This is iterative evaluation — you test performance every few epochs or after each training round to make sure learning is on track.

---

### 🧠 After Training

Once the model is fully trained, you perform a **final, structured evaluation** using a **test dataset** (or a held-out validation set).
This is where the full **Evaluation Methodology** applies:

1. **Define success criteria** — decide what good performance means (e.g., accuracy above 90%, low RMSE, fair distribution of outcomes, safe or relevant text outputs).
2. **Prepare a ground truth dataset** — curated or labeled examples with correct answers that serve as the gold standard.
3. **Run the trained model** — generate predictions or outputs for every example in your evaluation dataset.
4. **Compare predictions with ground truth** — measure correctness, closeness, or quality using metrics such as accuracy, precision, recall, F1 score, BLEU, ROUGE, or MSE.
5. **Aggregate and interpret results** — look at average scores, identify common failure cases, and measure consistency across data subsets.

Finally, **report findings in a clear and interpretable format**, such as:

* **Charts** (e.g., accuracy or loss over epochs).
* **Leaderboards** (for comparing multiple models or configurations).
* **Confusion matrices** (to visualize which classes are often misclassified).

These visual summaries make it much easier to spot where the model performs well or struggles — for example, a confusion matrix might reveal that “cats” are often mistaken for “foxes.”

---

### ⚙️ After Deployment (Ongoing Evaluation)

Evaluation doesn’t stop after model release.
You continue monitoring **real-world performance**, especially when:

* Input data distributions shift (known as *data drift*).
* Model outputs start deviating from expected behavior.
* New feedback, user interactions, or ground truth labels become available.

At this stage, dashboards, monitoring tools, and periodic retraining help keep the model reliable.

---

### ✅ In Summary

> Evaluation is not a one-time activity.
> It starts **during training**, becomes formalized **after training**, and continues **after deployment**.

It involves:

* Measuring how well predictions align with ground truth.
* Summarizing results in interpretable visual formats (charts, confusion matrices, leaderboards).
* Using those insights to refine, retrain, or replace models over time.

Together, these steps ensure that your model’s performance is not just accurate once — but consistently interpretable, reliable, and aligned with real-world expectations.
