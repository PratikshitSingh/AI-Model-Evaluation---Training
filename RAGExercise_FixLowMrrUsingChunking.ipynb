# 🧩 Exercise: Fix Low MRR Using Chunking

In this exercise, you will improve ranking quality (**MRR**) by applying **intent-based chunking** to an authoritative policy document.

You are **not changing the model**, **not changing Top-K**, and **not changing the LLM**.
You will fix the problem **only by restructuring the data**.

---

## Background

You have already observed that:

* Recall@K = 1
* MRR = 0.25 (the correct document appears at rank 4)

This means the correct document **is retrieved**, but it is **ranked too low to matter**.

The root cause is that the authoritative policy document:

* covers multiple tools
* contains the correct answer in only one sentence
* is semantically broader than the user’s question

Dense retrievers embed **average meaning**, so important details can get diluted.

---

## Goal

Improve MRR by making the **Slack access requirement** the *primary semantic signal*.

You will do this by **chunking the authoritative policy into smaller, focused pieces**.

---

## What you are allowed to change

* You may add **new chunks** to the existing Weaviate class (`Document5`) - leave the existing "polluted" ones alone
* You may reuse the same embedding model and instruction
* You may run retrieval as many times as needed

---

## What you may NOT change

* Do not delete existing documents
* Do not change the embedding model
* Do not change Top-K
* Do not add re-ranking or filtering logic
* Do not modify the LLM prompt

This exercise is about **data modeling**, not algorithms.

---

## Guidelines for Chunking

Use these rules to design your chunks:

1. **One requirement per chunk**

   * Each chunk should express exactly one access rule

2. **Make the query topic dominant**

   * If the chunk is about Slack, Slack should be central to the text

3. **Avoid multi-topic chunks**

   * Do not mix Slack, GitHub, and finance rules in the same chunk

4. **Keep chunks short but complete**

   * Short enough to stay focused
   * Complete enough to stand alone as a policy statement

5. **Preserve authority**

   * The chunk should clearly read like an official requirement

---

## Your task

1. Identify the authoritative policy document that contains the Slack access rule
2. Decide how many chunks it should be split into
3. Write the chunk text so that each chunk represents one clear requirement
4. Embed and add those chunks to the `Document3` class
5. Re-run retrieval for the question:

```
What do I need to access Slack?
```

---

## What to observe

After chunking, examine:

* Where does the correct chunk appear in the ranked results?
* Did MRR improve?
* Did polluted or generic documents drop in rank?

You should see the correct requirement move **earlier in the list**, often into the top 1–2 positions.
