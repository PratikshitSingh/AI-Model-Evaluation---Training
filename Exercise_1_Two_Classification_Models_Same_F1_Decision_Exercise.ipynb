# 🧪 Exercise 1 — Two Models, Same F1: Which One Would You Ship?

## Context
In this exercise, you are **not training a model** and **not writing evaluation code**.

Instead, you are playing the role of a **technical decision-maker** reviewing model evaluation results **before deployment**.

Two classification models have been evaluated on the **same test dataset**. Both models report:

- Identical accuracy
- Identical macro F1 score

However, their **confusion matrices differ**.

Your task is to **analyze the error patterns** and decide which model you would deploy in a real system.

---

## Labels Used in This System

- `billing_issue`
- `account_access`
- `technical_bug`
- `subscription_question`

Assume this is an **automated support ticket routing system** where predictions may trigger downstream workflows.


## 🔢 Provided Evaluation Results (Read-Only)

The following confusion matrices were generated using the same test set.

**Do not modify this cell.**

Your goal is to interpret the results — not compute them.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

labels = [
    "billing_issue",
    "account_access",
    "technical_bug",
    "subscription_question"
]

# Confusion Matrix for Model A
cm_model_a = np.array([
    [3, 0, 0, 0],
    [0, 1, 2, 0],
    [0, 0, 2, 0],
    [0, 0, 0, 2]
])

# Confusion Matrix for Model B
cm_model_b = np.array([
    [3, 0, 0, 0],
    [0, 2, 0, 1],
    [0, 0, 2, 0],
    [1, 0, 0, 1]
])

def plot_confusion_matrix(cm, title):
    plt.figure(figsize=(6, 5))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=labels,
        yticklabels=labels
    )
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(title)
    plt.show()

plot_confusion_matrix(cm_model_a, "Model A — Confusion Matrix")
plot_confusion_matrix(cm_model_b, "Model B — Confusion Matrix")

print("Model A — Accuracy: 0.75 | Macro F1: 0.73")
print("Model B — Accuracy: 0.75 | Macro F1: 0.73")
```
